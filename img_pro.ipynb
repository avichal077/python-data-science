```python
# load impage from url
import requests  

def laod_image(url):
    return Image.open(requests.get(url,stream = True).raw)
```

```python
url = "https://wikimedia.org/static/images/project-logos/enwiki-2x.png"
im = laod_image(url)
im
```

```python
imp_path.transpose(Image.FLIP_TOP_BOTTOM)
```

```python
print(imp_path.height,imp_path.width,imp_path.mode)
```

```python
imp_path.resize((300,300))
```

```python
imp_path.rotate(45,fillcolor='white')
```

```python
imp_path.rotate(90,expand=True)
```

```python
w,h = imp_path.width,imp_path.height
imp = imp_path.resize((w//2,h//2))
imp
```

```python
# enhancing color 
en = ImageEnhance.Color(imp)
```

```python
for i in range (-5,6):
    en.enhance(i).save(f"imp/imp_enhance_{i}.png")
```

```python
en.degenerate # same as en.enhance(0)
```

```python
imp_path.resize((w*2,h*2)).size
```

```python
# image drawing
imp = imp_path.resize((w//2,h//2))
dr = ImageDraw.Draw(imp)
dr.circle((175,140),radius = 80,fill = 'red')
dr.circle((175,140),radius = 100,outline = 'yellow',width= 10)
imp
```

```python
from PIL import ImageFont
font = ImageFont.truetype(r"c:\Windows\Fonts\AGENCYR.TTF",50)
dr.text((175,140),'target',fill = 'black',font=font,stroke_fill='white',stroke_width=5, anchor ='ms')
imp
```

```python
start_point = (0,imp.height//2)
end_point = (imp.width , imp.height //2)
dr.line(xy=(start_point,end_point),fill = 'black' ,width = 5)
dr.line(xy=((imp.width//2,0),(imp.width//2,imp.height)),fill = 'black',width = 5)
imp
```

```python
imp.save('imp/generated.png')
```
